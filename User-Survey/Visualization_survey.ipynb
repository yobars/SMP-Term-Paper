```python
plt.figure(figsize=(15, 6))
# Plot age distribution
plt.subplot(2, 4, 5)
df['Age'].value_counts().sort_index().plot(kind='bar', color='purple')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of People')

# Plot gender distribution
plt.subplot(2, 4, 6)
df['Gender'].value_counts().plot(kind='bar', color='pink')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of People')

# Plot profession distribution
plt.subplot(2, 4, 7)
df['Profession'].value_counts().plot(kind='bar', color='brown')
plt.title('Profession Distribution')
plt.xlabel('Profession')
plt.ylabel('Number of People')

plt.tight_layout()
plt.show()
```

```python
platforms = ["Daraz", "Rokomari", "Chaldal", "Bikroy"]
colors = ['blue', 'green', 'orange', 'red']

plt.figure(figsize=(10, 5))

for i, platform in enumerate(platforms):
    plt.subplot(1, 4, i + 1)
    usage_counts = df[f"{platform.lower()}Usage"].value_counts().sort_index()
    usage_counts.plot(kind='bar', color=colors[i])
    plt.title(f'{platform} Usage')
    plt.xlabel('Usage (Yes/No)')
    plt.ylabel('Number of People')

plt.tight_layout()
plt.show()
```

```python
platforms = ["Daraz", "Rokomari", "Chaldal", "Bikroy"]
usability_columns = [f"{platform.lower()}Navigation" for platform in platforms] + \
                    [f"{platform.lower()}InfoUnderstand" for platform in platforms] + \
                    [f"{platform.lower()}Checkout" for platform in platforms]

# Calculate overall usability score for each platform for each respondent
for platform in platforms:
    df[f'Usability{platform}'] = df[[f'{platform.lower()}Navigation',
                                     f'{platform.lower()}InfoUnderstand',
                                     f'{platform.lower()}Checkout']].mean(axis=1)

# Calculate the average usability score for each platform
average_usability_platforms = df[[f'Usability{platform}' for platform in platforms]].mean()

# Plot a bar chart for the average usability scores for each platform
plt.figure(figsize=(8, 4))
plt.bar(platforms, average_usability_platforms, color=['blue', 'green', 'orange', 'red'])
plt.title('Average Usability Scores for Platforms')
plt.xlabel('Platform')
plt.ylabel('Average Usability Score')
plt.ylim(0, 5)  # Set y-axis limit based on your rating scale
plt.show()
```
